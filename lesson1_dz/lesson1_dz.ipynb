Задача 1

Подбрасываются три игральных кубика. Оцените вероятность того, что сумма значений этих кубиков не будет превышать 6.

Подсказка: оценкой вероятности может служить, например, относительная частота при большом числе опытов. Попробуйте реализовать наибольшее число опытов, которое позволяют ваш компьютер и/или ваше терпение.

```python
def cube_throw(a):
    """
    a -  number of edges
    return random number in range[1; a]
    """
    return(randint(1, a))
N = 100000
MAX_VALUE = 6
envents_count = 0
for i in range(N):
    event_value = [cube_throw(MAX_VALUE) for i in range(3)]    
    if sum(event_value) <= 6:
        envents_count += 1
#относительная частота будет равна отношению events_count / N
print(envents_count / N)
```

Вероятность такого событя полученая опытным путем будет  приблизительно 9%

Задача 2

Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

Количество сочетаний  4-х карт из колоды(52 карты) будет $C_{52}^4$.<br>
Количество вариантов расположить 1 туза в 4х картах будет $C_4^1$. Дальше нам надо их перемножить, хотя эта часть совершенно не очевидна, и я делаю по сути просто по аналогии с уроком. Я понимаю почему это происходит с вероятностями, но почему в данном случае так нужно поступать с количеством сочетаний совершенно непонятино. Ну и в общем вы и не объясняли на вебинаре.



```python
def combinations(n: int, k: int) -> int:
    """Число сочетаний.
    """
    
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
    
def arrangements(n: int, k: int) -> int:
    """Число размещений.
    """
    
    return np.math.factorial(n) // np.math.factorial(n - k)
```

```python
print("количество возможностей будет равно: ", combinations(52, 4) * combinations(4, 1))
```

Задача 4

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Формула условной вероятности $$P(A|B) = \frac{P(A \cdot B)}{P(B)}$$ <br>
$P(B) = 1/2$ так как 3 числа больше 3 из 6.<br>
$P(A \cdot B) = 1/2 * 2/3 = 2/3$ так как попав в 3 верхних 4,5,6 две из них будет четными.<br>
$P(A|B) = 2/3$<br>
$P(A) = 1/2$ <br>
таким образом события получаются зависимы так как наступление события B увелиичвает вероятность события А

Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?

Подсказка: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Событие А - мы заболели, событие B - тест положительный.
По формуле Байеса  $$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$
$P(A) = 1/1000$ так как болеет всего 1 из тысячи<br>
$P(B|A) = 99/100$ так как вероятсность того что тест даст нам положительный результат если мы больны в 99% сулчаев.<br>
формула полной вероятности в данном случае $P(B) = P(B|A) \cdot P(A) + P(B| \overline{A}) \cdot P(\overline{A})$<br>
$P(B|\overline{A}) = 1/100$ вероятность того что мы получили положительный тест и при условии что мы не заболели.<br>
$P(\overline{A})  = 999/1000$ вероятность впринципене незаболеть.<br>
подставляем все в формулу полной вероятности.

```python

print("веротность быть больным получается: ", 99/100/1000/(99/100/1000 + 999/1000/100))
```
